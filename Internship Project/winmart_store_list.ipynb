```python
import requests
from bs4 import BeautifulSoup

URL = "https://winmart.vn/info/danh-sach-cua-hang"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
meta_attributes = []

for meta_tag in soup.find_all("meta"):
    if meta_tag.get("property"):
        meta_attributes.append(("property", meta_tag["property"]))
    elif meta_tag.get("name"):
        meta_attributes.append(("name", meta_tag["name"]))

# Print each found property and name
print("Meta properties and names:")
for attr_type, attr_value in meta_attributes:
    print(f"{attr_type}: {attr_value}")
```

```python
import pandas as pd

data = response.json()

store_list = []

for district in data["data"]:
  district_name = district["districtName"]

  for ward in district['wardStores']:
        ward_name = ward['wardName']

        # Check if there are stores in the ward
        if ward['stores']:

            for store in ward['stores']:
                store_name = store['storeName']
                store_address = store['officeAddress']

                store_list.append({
                    "District": district_name,
                    "Ward": ward_name,
                    "Store Name": store_name,
                    "Store Address": store_address
                })

df = pd.DataFrame(store_list)
df.to_excel("HCI.xlsx", index=False)
```
